## L1, L2 Regularization </h1>

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
df.head(3)
```

```python
X = df.drop(['target'], axis = 1)
y = df['target']
```

```python
X.sample(5)
```

```python
y.sample(5)
```

### Training using Simple linear regression

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Training using Lasso regression

```python
lasso_model = Lasso(alpha = 1.0) # Lasso = linear regression with MSE loss + L1 regularization (drives some coefficients to zero)
lasso_model.fit(X_train, y_train)
lasso_model.score(X_test, y_test)
```

### Training using Ridge regression

```python
ridge_model = Ridge(alpha = 1.0) # Ridge  = linear regression with MSE loss + L2 regularization (shrinks coefficients but doesn’t make them zero)
ridge_model.fit(X_train, y_train)
ridge_model.score(X_test, y_test)
```
